## Problem Statement

- Customer churn has a direct financial impact on telecom companies.
- Acquiring a new customer typically costs **5–7 times more** than retaining an existing one.
- Therefore, predicting churn early helps companies reduce revenue loss and design targeted retention strategies.

## Customer Churn Analysis

- Customer churn analysis involves understanding why customers stop using a company's product or service.
- **Churn:** When a customer leaves or stops using the service.

# Split data into X and y

```python
x = data.drop("Churn", axis = 1)
y = data["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Feature Engineering
* Convert categorical → numerical (Encoding)
* Label Encoding / One-Hot Encoding
* Drop unnecessary columns (like customerID)

```python
data.head()
```

## Handling Categorical Columns
A) Yes/No Columns → Convert to 1/0

```python
binary_cols = ["Partner","Dependents","PhoneService","PaperlessBilling"]

for col in binary_cols:
    x_train[col] = x_train[col].map({"Yes":1, "No":0})
    x_test[col]= x_test[col].map({"Yes":1, "No":0})
    
x_train.head(2)
```

## Converting Y into binary

```python
y_train = y_train.map({"No":0, "Yes":1})
y_test = y_test.map({"No":0, "Yes":1})
```

## 🔹 B) One-Hot Encoding (Multi-category columns)

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)
```

```python
x_train, x_test = x_train.align(x_test, join="left", axis=1, fill_value=0)

# It makes sure:
# Both train and test have exact same columns
# If a column is missing in test → it fills with 0
# align() = 👉 “Make train and test columns exactly same.”
```

### Convert All Boolean Columns to 0/1

```python
x_train = x_train.astype(int)
x_test = x_test.astype(int)
```

```python
x_train.head(3)
```

## Scaling Data with StandardScaler

```python
from sklearn.preprocessing import StandardScaler

num_cols = ["tenure", "MonthlyCharges","TotalCharges"]

scaler = StandardScaler()

x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

# Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight="balanced")
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score, confusion_matrix
print('Accuracy:', accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = "d", cmap = "Reds")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()


# fmt='d' → Show numbers as integers
# cmap='Reds' → Use red color theme
```

## 🔹 Before (Without Balancing)
* Accuracy ≈ 79%
* Recall (Yes) ≈ 52%

## 🔹 After class_weight="balanced"
* Accuracy ≈ 73% ⬇
* Recall (Yes) ≈ 79% ⬆🔥

## Insights:
Model is now :  
* Caticing 79% churn customers
* But overall accuracy is reduced(normal in imbalanced handling).
* Earlier:  
 ❌ Model was missing many churn customers.
* Now:  
✅ Model is catching most churn customers.
* For churn prediction, Recall (Yes) is more important than Accuracy.


# ⭐ Final Conclusion
* Balanced model is better for churn problem
because detecting churn customers is the main goal.

## Cross Validation for Logistic Regression

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, x_train, y_train, cv = 5, scoring = "recall")
print(scores)
print(scores.mean())
```

# Insight
* Using 5-fold cross-validation, the model achieved an average recall of 80% for churn customers, indicating stable and strong detection performance.

## Hyperparameter Tuning for Logistic

```python
from sklearn.model_selection import GridSearchCV
#from sklearn.linear_model import LogisticRegression

param_grid = {
    "C": [0.01, 0.1, 1, 10],
    "penalty": ["l2"],
    "solver": ["liblinear"]
}

grid = GridSearchCV(
    LogisticRegression(class_weight="balanced"),
    param_grid,
    cv=5,
    scoring="recall"
)

grid.fit(x_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Recall:", grid.best_score_)
```

# Model Camparison

```python
# step 1 : Import Models
#from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# step 2 : Create Model Dictionary
models = {
    "Logistic (Balanced)": LogisticRegression(class_weight="balanced"),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(class_weight="balanced"),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}
```

```python
# step 3 : Train & compare
from sklearn.metrics import recall_score
results = []

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    recall = recall_score(y_test, y_pred)
    results.append([name, recall])

df_results = pd.DataFrame(results, columns=["Model", "Recall (Yes)"])
df_results.sort_values(by="Recall (Yes)", ascending=False)
```

# 📊 Recall (Yes) Comparison
* Logistic (Balanced) → 0.791 🔥
* SVM → 0.778
* KNN → 0.516
* Decision Tree → 0.505
* Random Forest → 0.452

```python
from sklearn.metrics import precision_score
results = []

for name, model in models.items():
    
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    results.append([name, acc, precision, recall, f1])

df_results = pd.DataFrame(results, columns=[
    "Model",
    "Accuracy",
    "Precision",
    "Recall",
    "F1 Score"
])

df_results.sort_values(by="Recall", ascending=False)
```

# Model Comparison Insights
*  Logistic Regression (Balanced) is best  
*  SVM is second best  
*  Tree models are performing poorly for churn detection  
* Since churn problem needs high recall,  
👉 Logistic (Balanced) is currently our best model.

### ROC-AUC Curve
ROC-AUC measures the model’s ability to distinguish between churn and non-churn customers.
Higher AUC indicates better classification performance.

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_prob = model.predict_proba(x_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {auc_score:.3f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

### Precision-Recall Curve
This curve is useful for imbalanced datasets and focuses on the model’s ability to detect churn customers.

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```

## HyperParameterTunning for SVM

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C" : [0.1, 1, 10],
    "kernel" : ["linear","rbf"],
    "gamma" : ["sclae", "auto"]
}

grid = GridSearchCV(
    SVC(class_weight="balanced"),
    param_grid,
    cv = 5,
    scoring = "recall"
)

grid.fit(x_train, y_train)

print("Best Parameters :", grid.best_params_)
print("Best Recall :", grid.best_score_)
```

## Hyperparameter Tuning Comparison

| Model | Default Recall | Tuned Recall | Improvement | Final Decision |
|-------|----------------|--------------|-------------|----------------|
| Logistic (Balanced) | 0.79 | 0.80 | Slight Increase | Good but not best |
| SVM (Balanced) | 0.78 | 0.83 | Significant Increase | Selected |

## 📌 Insights
* After tuning, SVM recall improved significantly (~83%), meaning it detects more churn customers.
* Logistic Regression tuning gave stable performance (~80% recall) but slightly lower than SVM.
* Hyperparameter tuning clearly improved model performance compared to default settings.
* Final decision: Tuned SVM performs best for churn detection. 🚀
